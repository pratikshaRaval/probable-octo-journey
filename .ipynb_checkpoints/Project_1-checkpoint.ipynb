    Project Objective: 1) From 4 options, take user input on menu choice(usrIn) then a natural 
                          number(usrNo) 
                       2) Based on the menu input, perform 1 - Fibonacci series on the given userNo 
                          (Generate a fibonacci series till the user given number) 
                       3) For the given number, print "Fizz" if its a multiple of 3, "Buzz" if a 
                          multiple of 5 and "FizzBuzz" if a multiple of 15 
                       4) Generate random numbers list of user given length in the range [-50,50] 
                       5) Press '0' for exiting.

```python
import random 

##### Option 1 Function: Fibonacci Series till the usrNo (user Input) 

def funcFib():
    #Check for the usrNo, if it is numeric and a natural number then only proceed

    while True:             #An infinite loop till the number entered is numeric
        usrNo = input("Give the number (n):")
        if usrNo.isnumeric() and int(usrNo)!=0:
                break       #indent 'coz break is under the if control statement condition
        else: 
                print("\nInvalid number, please put only natural numbers")
            
    no1,no2 = 0,1 
    count = 0
    
    if int(usrNo)==1:
        print (usrNo)
    else:
        while count<int(usrNo):
            print (no1)
            nNo = no1 + no2
            no1 = no2
            no2 = nNo
            count += 1


##### Option 2 Function:  Till the usrNo (user Input), print "Fizz" for multiples of 3,
#####"Buzz" for multiples of 5, "FizzBuzz" for multiples of 15 
##### After # I can add whatever I want so number of # doesn't matter

def funcMul():
    while True:
        usrNo = input("Give the number (n):")
        if usrNo.isnumeric() and int(usrNo)!=0: # If we put 0 then 0 is always the answer
            break                               #,thus the zero check  
            
        else: 
            print("\nInvalid number, please put only natural numbers")
        
    for i in range(1,int(usrNo),1):
            if  not i%15:
                print ("FizzBuzz")
            elif not i%5:
                print ("Buzz")
            elif not i%3:
                print ("Fizz")
            else:
                print (i)

##### Option 3 Function: Prime numbers till the usrNo (user Input) 
def funcPrm():
    while True:
        usrNo = input("Give the number (n):")
        if usrNo.isnumeric() and int(usrNo)!=0:
            break
        else: 
            print("\nInvalid number, please put only natural numbers")

    start = 1
    end = int(usrNo) #usrNo is originally a string so type converting it
                                #to int and then adding 1
    for val in range(start, end + 1): 
        if val > 1: 
            for n in range(2, val): 
                if (val % n) == 0: 
                    break
            else:
                print(val)
    
##### Option 4 Function: Generate random numbers list of length usrNo in [-50 to 50] 

def funcRnd():
    while True:
        usrNo = input("Give the number (n):")
        if usrNo.isnumeric() and int(usrNo)!=0 and int(usrNo)<=101:#Since the range is 100,101
            break                                #to include the 50th term in positive integer
        else: 
            print("\nInvalid number, please put only natural numbers")
            
   
    lst = []

    for x in range(int(usrNo)):
        lst.append(random.randint(-50,50))

    print (lst)
    
##### Option 0 Function: Exit the program 
def funcExit():
    print("\nExiting...") #just printing Exit

"""
Creating a dictonary for the menu options. Dictonary Keys as the option number and function names
as their values.   

"""
dictMenu = {'1':funcFib,'2':funcMul,'3': funcPrm,'4':funcRnd,'0': funcExit}

"""
Declaring a choice variable to control the loop which needs to be infinite unless user inputs
'0' from the menu. 
"""
goodChoice = True
    
while goodChoice:
    
    #Just print a print and a menu for UI
    print ("_"*90 + "\nOption 1: Fibonacci Series\nOption 2: Multiples of 3, 5 and both\nOption 3: Print the prime numbers \nOption 4: Print random number list of length n")
        
    #User Input validation for whole numbers
    while True:
        usrIn = input("Please give an option from the above menu or press 0 to exit: ")
        if usrIn.isnumeric():
            break
        else:
            print ("\nPlease select the correct menu option")
    
#We are checking if the keys in dictMenu match the value of usrIn
    if usrIn in dictMenu.keys():
        if int(usrIn)==0: #0 is for exiting so has a different behavior
            dictMenu[usrIn]() #referencing the function () via dictMenu names
            break #come out of the loop and do not repeat the menu options
        else:
            dictMenu[usrIn]()
    else:
            print("\nPlease give proper choice") #Tell user to re-enter the value for menu if its
                                               # not in the ditcMenu [1,2,3,4,0]
  
       


 
```
